## Simple

```python
# Import Libraries

import requests
from bs4 import BeautifulSoup as bs

URL = "https://www.scrapethissite.com/pages/simple/"

# Request the website
response = requests.get(URL)

# Response
if(response.status_code == 200):
    print("Success")
    print("Response Code - %d" % (response.status_code))
else:
    print("Failed")
    print("Response Code - %d" % (response.status_code))

# Assign Parser
page = bs(response.text, "html.parser")

# Parse
allCountries = page.find_all("div", class_="country")
print("Total Count - %d" % (len(allCountries)))

count = 0

for country in allCountries:
	count = count + 1
	name = country.find("h3", class_="country-name")
	print("%d. %s" % (count, name.text.strip()))
```

## Pagination & Request Type

```python
# Import Libraries

import requests
from bs4 import BeautifulSoup as bs

URL = "https://www.scrapethissite.com/pages/forms/"

# Request the website
response = requests.get(URL, params=dict(
    page_num=1,
    per_page=25
))

# Response
if(response.status_code == 200):
    print("Success")
    print("Response Code - %d" % (response.status_code))
else:
    print("Failed")
    print("Response Code - %d" % (response.status_code))

# Assign Parser
page = bs(response.text, "html.parser")

# Parse
allTeams = page.find_all("tr", class_="team")
print("Total Count - %d" % (len(allTeams)))

count = 0

for team in allTeams:
	count = count + 1
	name = team.find("td", class_="name").text.strip()
	year = team.find("td", class_="year").text.strip()
	wins = team.find("td", class_="wins").text.strip()
	losses = team.find("td", class_="losses").text.strip()
	print("%d. Team %s has won %s matches and lost %s matches in the year %s" % (count, name, wins, losses, year))
```

```python
# How we normally Request to a site

URL = "http://0722-49-204-139-210.ngrok.io"

# Request Page
response = requests.get(URL)

# Response
if(response.status_code == 200):
    print("Success", end="\n"*2)
    print("Response Code - %d" % (response.status_code), end="\n"*2)
else:
    print("Failed", end="\n"*2)
    print("Response Code - %d" % (response.status_code), end="\n"*2)

print(response)
print("\n")
print(response.json())
```

```python
# How to spoof an user-agent to a site

URL = "http://0722-49-204-139-210.ngrok.io"
customHeaders = {}
customHeaders["User-Agent"] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"

# Request Page
response = requests.get(URL, headers=customHeaders)

# Response
if(response.status_code == 200):
    print("Success", end="\n"*2)
    print("Response Code - %d" % (response.status_code), end="\n"*2)
else:
    print("Failed", end="\n"*2)
    print("Response Code - %d" % (response.status_code), end="\n"*2)

result = response.json()
print(result, end="\n"*2)
print("User-Agent is %s" % (result["User-Agent"]), end="\n"*2)
print("IP Address is %s" % (result["ipaddress"]))
```
